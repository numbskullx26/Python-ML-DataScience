# Difference Between Random Forests and Decision Trees

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = "C:\\Users\\Indrajeet Sarkar\\Desktop\\Python-ML\\car_evaluation.csv"



df = pd.read_csv(data, header=None)
```

```python
col_names=['buying','maint','doors','persons','lug_boot','safety','class']

df.columns=col_names

col_names
```

```python
for col in col_names:
    print(df[col].value_counts())
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop(['class'], axis=1) #  .drop()is used to remove or drop out a column from the dataset.Independent: as these 6 columns do no depend on anything.
```

```python
y = df['class'] #Dependent values: as it depends on the other 6 columns.
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33, random_state= 42)
# here we include 4 variables for training and testing. 
#test_size = 0.33 means 33% of values of the datatset will go testing and the rest 67% wil go for training
```

```python
import category_encoders as ce #encoding
#for changing the categorial values to numbers
encoder =  ce.OrdinalEncoder(cols=['buying','maint','doors','persons','lug_boot','safety'])
# many methods are used for performing this operation. One of them is Category Encoders
#Here we are applying it on the columns 

x_train = encoder.fit_transform(x_train)
#both fit_transform and transform are methods for transforming categorial values into numbers
x_test= encoder.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

#instantiate the classifier

rfc = RandomForestClassifier(random_state=0)

#fit the model on the training data 

rfc.fit(x_train,y_train)

#predict the test results

y_pred =  rfc.predict(x_test)

#check accuracy score

from sklearn.metrics import accuracy_score

print("Model accuracy score with 10 decision-trees : {0:0.4f}" .format(accuracy_score(y_test,y_pred)))
```
