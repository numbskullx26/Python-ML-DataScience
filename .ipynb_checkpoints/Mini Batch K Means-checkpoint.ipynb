# Mini-Batch K-Means clustering digits classification

```python
from sklearn.datasets import load_digits
digits = load_digits()
data = digits.data
```

```python
data
```

```python
data = data/255
```

```python
from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(n_clusters=15, batch_size=6, max_iter=10) #making the model
kmeans.fit(data) #fitting the model on the 'data'
predictions = kmeans.predict(data)
```

```python
import numpy as np
# import matplotlib.pyplot as plt
# %matplotlib inline 
for i in range(0,15):
    
    row = np.where(predictions==i)[0] #.where() is basically checking a condition which is used for filtering. [0]is used for picking up only 0 values.
    num = row.shape[0]
    rows = np.floor(num/10.)
    
    print("Cluster no: "+ str(i))
    print(str(num)+ " items")
    plt.figure(figsize=(10,7))
    for k in range(0,num):
        plt.subplot(rows+1 , 10, k+1)
        image = data[rows[k], ]
        image = image.reshape(8,8)
        plt.imshow(image,cmap='cividis')
        plt.axis('off')
    plt.show()
```
