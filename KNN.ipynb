```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = fruits[['mass','width','height']] #independentvariables
Y = fruits['fruit_label'] #Target Variable || Dependent Variable
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=0)
```

```python
#dataset
# Independent(X), Dependent(Y)
# Independent(X_train,X_test),Dependent(Y_train,Y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
X_train.head()
```

```python
knn.fit(X_train,Y_train) #fitting the knn data model on our train data
```

```python
knn.score(X_test,Y_test) #showing us how accurate our dataset actually is.
```
