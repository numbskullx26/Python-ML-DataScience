# Exception Handling 

```python
l1 = list(range(0,10))
```

```python
l1
```

```python
l1[6]
```

```python
l1[7]
```

```python
l1[10]
```

```python
try:
    l1[10] #error
except:
    print("Please enter a valid index")
```

```python
p = int(input(f"Please enter a valid index to retrieve the value from the list: {l1}"))

try:
    print(f"The index you entered is {p} and the value at that index in the list is : {l1[p]}")

except:
    print(f"Please enter a valid index between 0 and {len(l1)-1}")
    
else:
    print("This statement will be printed under else") # this is executed if no error/exception is encountered
    
finally:
    print("This statement will be printed under finally")# this is always executed
```
